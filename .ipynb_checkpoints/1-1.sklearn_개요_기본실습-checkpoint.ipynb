```python
import numpy as np
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.arange(1,30,10).reshape(-1,1) # -1 : 알아서 알맞게 설정
y = 2*X + 1
```

```python
X
```

```python
sns.regplot(x=X, y=y)
```

```python
model_lr = LinearRegression()
model_lr
```

```python
model_lr.fit(X,y)
```

```python
model_lr.predict([[50],[70],[200]])
```
