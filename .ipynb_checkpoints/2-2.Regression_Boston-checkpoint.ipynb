```python
X_df = bh_df.drop('price', axis=1)
y = bh_df['price']
```

```python
X_df.head()
```

```python
y.head()
```

```python
X = X_df.values
print(type(X))
X[:2]
```

```python
from sklearn.model_selection import KFold

num_split = 5

# kf = KFold(n_splits=num_split, shuffle=True, random_state=40)
kf = KFold(n_splits=num_split)

tot_MSE = 0.0

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model_lr = LinearRegression()
    model_lr.fit(X_train, y_train)
    y_pred = model_lr.predict(X_test)

    tot_MSE += mean_squared_error(y_test, y_pred)

MSE = tot_MSE / num_split
RMSE = np.sqrt(MSE)
print('MSE : {0:.5f}'.format(MSE))
print('RMSE : {0:.5f}'.format(RMSE))
```

```python
print('회귀계수: ', np.round(model_lr.coef_, 2))
print('절편 : ', model_lr.intercept_)
```

```python
X = bh_df.loc[:, useful_feature_list].values
y = bh_df.iloc[:,-1].values
```

```python
num_split = 5

kf = KFold(n_splits=num_split, shuffle=True, random_state=40)

tot_MSE = 0.0

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model_lr = LinearRegression()

    model_lr.fit(X_train, y_train)

    y_pred = model_lr.predict(X_test)

    tot_MSE += mean_squared_error(y_test, y_pred)

MSE = tot_MSE / num_split
RMSE = np.sqrt(MSE)
print('MSE : {0:.5f}'.format(MSE))
print('RMSE : {0:.5f}'.format(RMSE))
```

```python
from sklearn.model_selection import cross_val_score

model_lr = LinearRegression()

neg_mse_scores = cross_val_score(model_lr, X, y, scoring='neg_mean_squared_error', cv=5)
mse = (-1 * neg_mse_scores).mean()
rmse = np.sqrt(mse.mean())

print('5 folds의 개별 Negative MSE scores: ', np.round(neg_mse_scores,2))
print('5 folds의 평균 MSE: ', np.round(mse,5))
print('5 folds의 평균 RMSE: {0:.5f}'.format(rmse))

neg_root_mse_scores = cross_val_score(model_lr, X, y, scoring='neg_root_mean_squared_error', cv=5)
print('5 folds의 평균 RMSE : ', -1 * neg_root_mse_scores.mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=2024)

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

num_split = 5

kf = KFold(n_splits=num_split, shuffle = True, random_state=2024)

tot_MSE = 0.0

for train_index, test_index in kf.split(X_train):
    X_train_kf, X_test_kf = X_train[train_index], X_train[test_index]
    y_train_kf, y_test_kf = y_train[train_index], y_train[test_index]

    model_lr4 = LinearRegression()
    model_lr4.fit(X_train_kf, y_train_kf)
    y_pred_kf = model_lr4.predict(X_test_kf)
    tot_MSE += mean_squared_error(y_test_kf, y_pred_kf)

MSE = tot_MSE / num_split
RMSE = np.sqrt(MSE)
print('MSE : {0:.5f}'.format(MSE))
print('RMSE : {0:.5f}'.format(RMSE))
```
