```python
df = pd.read_csv('clean_house_df.csv')
```

```python
df.head()
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1}, inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1}, inplace = True)
```

```python
df.info()
```

```python
cat_cols
```

```python
df = pd.get_dummies(df, columns=['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop('RainTomorrow', axis=1)
```

```python
y = df['RainTomorrow']
```

```python
X.head()
```

```python
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=2023)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
```

```python
y_pred = dt_clf.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
print('결정 트리 (Decision Tree) Accuracy :', score)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)

y_pred = rf_clf.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
print('랜덤 포레스트 Accuracy :', score)
```

```python
print(classification_report(y_test, y_pred))
```
